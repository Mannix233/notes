Softmax 回归（用于分类问题）核心逻辑：

多输出： 模型的输出个数等于类别的个数（预测每个类别的得分）。

Softmax 函数： 负责把模型的得分，转换成加起来等于 1 的概率值。

交叉熵损失（CrossEntropyLoss）： 分类问题专属的计分板，用来计算预测概率和真实答案之间的误差。

# 损失函数可视化：3种“教练”的脾气
### 核心规则（死盯两根线）
- 横坐标（左右）：偏离马路中心的距离 = **你错得有多离谱**（0=完美猜中标准答案）
- 蓝线（Loss）：罚款金额 → 错得越远，罚款越狠
- 橙线（Gradient）：教练拽方向盘的力度 → 模型纠正错误的“力气”
- 绿线忽略！只看蓝+橙！

```python
import numpy as np
import matplotlib.pyplot as plt

# === 关键配置：适配Jupyter，中文显示+图表风格 ===
plt.rcParams['font.sans-serif'] = ['SimHei']  # Windows中文，Mac换['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False    # 负号正常显示
plt.rcParams['figure.facecolor'] = 'white'    # Jupyter白底

# === 1. 定义3种损失+梯度（完全对应你的“教练”描述） ===
def l2_loss(x):
    """L2均方损失：蓝线=U型碗，罚款成倍增加"""
    return 0.5 * np.square(x)

def l2_gradient(x):
    """L2梯度：橙线=直线，远猛拽、近轻揉"""
    return x

def l1_loss(x):
    """L1绝对值损失：蓝线=V型漏斗，直上直下"""
    return np.abs(x)

def l1_gradient(x):
    """L1梯度：橙线=水平阶梯，机器人教练，力度不变"""
    grad = np.ones_like(x)
    grad[x < 0] = -1
    grad[x == 0] = np.nan  # 原点断层=教练乱晃画龙
    return grad

def huber_loss(x, delta=1.0):
    """Huber损失：蓝线=缝合怪，远直近圆"""
    abs_x = np.abs(x)
    return np.where(abs_x <= delta, 0.5*np.square(x), delta*(abs_x - 0.5*delta))

def huber_gradient(x, delta=1.0):
    """Huber梯度：橙线=聪明教练，远稳近柔"""
    abs_x = np.abs(x)
    return np.where(abs_x <= delta, x, delta*np.sign(x))

# === 2. 生成数据：x=偏离距离（错得有多离谱） ===
x = np.linspace(-5, 5, 1000)  # 偏离范围：-5（左偏）→ 0（正中）→ 5（右偏）

# === 3. 计算各损失/梯度值 ===
l2_loss_vals = l2_loss(x)
l2_grad_vals = l2_gradient(x)

l1_loss_vals = l1_loss(x)
l1_grad_vals = l1_gradient(x)

huber_loss_vals = huber_loss(x)
huber_grad_vals = huber_gradient(x)

# === 4. 绘制3张图（上下竖排，替代原来的并排） ===
# 3行1列，宽度12（适合Jupyter），高度15（竖排足够展示）
fig, axes = plt.subplots(3, 1, figsize=(12, 15))  

# --- 第一张（最上方）：L2均方损失（远猛拽，近轻揉）---
ax1 = axes[0]
ax1.plot(x, l2_loss_vals, '#1f77b4', linewidth=3, label='蓝线=罚款金额')
ax1.plot(x, l2_grad_vals, '#ff7f0e', linewidth=3, label='橙线=教练拽方向盘力度')
ax1.axhline(0, color='black', linestyle='--', alpha=0.5)  # 水平参考线
ax1.axvline(0, color='black', linestyle='--', alpha=0.5)  # 0点=马路正中间
ax1.set_title('L2教练：远了猛拽，近了轻揉\n（极端错误易“拽断方向盘”）', fontsize=14, fontweight='bold')
ax1.set_xlabel('偏离马路中心的距离（错得有多离谱）', fontsize=12)
ax1.set_ylabel('金额/力度', fontsize=12)
ax1.legend(loc='upper right', fontsize=10)
ax1.grid(alpha=0.3)
ax1.set_ylim(-5, 8)  # 统一y轴，方便对比

# --- 第二张（中间）：L1绝对值损失（机器人教练）---
ax2 = axes[1]
ax2.plot(x, l1_loss_vals, '#1f77b4', linewidth=3, label='蓝线=罚款金额')
ax2.plot(x, l1_grad_vals, '#ff7f0e', linewidth=3, label='橙线=教练拽方向盘力度')
ax2.axhline(0, color='black', linestyle='--', alpha=0.5)
ax2.axvline(0, color='black', linestyle='--', alpha=0.5)
ax2.set_title('L1教练：机器人，力度永远不变\n（终点附近“疯狂画龙”）', fontsize=14, fontweight='bold')
ax2.set_xlabel('偏离马路中心的距离（错得有多离谱）', fontsize=12)
ax2.set_ylabel('金额/力度', fontsize=12)
ax2.legend(loc='upper right', fontsize=10)
ax2.grid(alpha=0.3)
ax2.set_ylim(-5, 8)

# --- 第三张（最下方）：Huber损失（聪明教练）---
ax3 = axes[2]
ax3.plot(x, huber_loss_vals, '#1f77b4', linewidth=3, label='蓝线=罚款金额')
ax3.plot(x, huber_grad_vals, '#ff7f0e', linewidth=3, label='橙线=教练拽方向盘力度')
ax3.axhline(0, color='black', linestyle='--', alpha=0.5)
ax3.axvline(0, color='black', linestyle='--', alpha=0.5)
ax3.axvline(1, color='red', linestyle=':', alpha=0.7, label='分界点：delta=1')
ax3.axvline(-1, color='red', linestyle=':', alpha=0.7)
ax3.set_title('Huber教练：聪明缝合怪\n（远稳近柔，不拽断+不画龙）', fontsize=14, fontweight='bold')
ax3.set_xlabel('偏离马路中心的距离（错得有多离谱）', fontsize=12)
ax3.set_ylabel('金额/力度', fontsize=12)
ax3.legend(loc='upper right', fontsize=10)
ax3.grid(alpha=0.3)
ax3.set_ylim(-5, 8)

# 调整布局，避免竖排时文字重叠
plt.tight_layout()
plt.show()

# === 实战提示（对应你的调试场景） ===
print("💡 调试机器人时的判断技巧：")
print("1. 模型在正确答案附近抖来抖去 → 是L1教练在画龙，换L2/Huber！")
print("2. Loss突然变成NaN（崩溃）→ 是L2教练拽断方向盘，换Huber！")
```

### 核心记忆点（记死这3句话）
1. L2教练：远猛拽、近轻揉 → 日常常用，但极端错误易“梯度爆炸”（拽断方向盘）
2. L1教练：力度不变 → 抗异常值，但终点画龙（稳不住）
3. Huber教练：聪明缝合怪 → 远稳（像L1）、近柔（像L2）→ 首选！
